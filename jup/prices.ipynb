```python
import pandas as pd

USD_RATE = 4.0

prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25)
]

df = pd.DataFrame(prices, columns = ['month', 'pricePLN'])
df.set_index('month')
df['priceUSD'] = df['pricePLN'] * USD_RATE
df['priceUSD'].plot(
    title = 'Price of goods (USD)',
    color = 'red',
    linestyle = 'dashed',
    grid = True,
    xticks = range(2, 13, 2),
    xlabel = 'month'
)
```
